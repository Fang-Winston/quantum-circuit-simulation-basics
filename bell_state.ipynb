```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram

# 1. 建立電路 (Bell State)
qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
qc.measure_all() # 自動加上測量與對應的古典位元

# 2. 繪製電路
display(qc.draw('mpl'))

# 3. 初始化模擬器
simulator = AerSimulator()

# 4. 編譯並執行
compiled_circuit = transpile(qc, simulator)
job = simulator.run(compiled_circuit, shots=1024)
result = job.result()

# 5. 取得結果並畫圖
counts = result.get_counts()
print("測量結果:", counts)
plot_histogram(counts)
```

Measurement results show that the two qubits are entangled, with '00' and '11' outcomes roughly equal. 
This confirms the correct creation of a Bell state.
